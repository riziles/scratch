```python
from sympy import symbols, sin, cos,  sqrt, Eq, S, pi
from sympy import Integral, evalf
from sympy.plotting import plot_parametric, plot
```

```python
theta = symbols('theta')
r = sqrt(1 + ( 3 / pi ) * theta)
A = Integral( 0.5 * r ** 2 , (theta , 0,  pi))
display ( Eq(S('r'), r, evaluate=False) )
display ( Eq(S('A'), A, evaluate=False) )
display(A.doit())
plot_parametric((r*cos(theta), r*sin(theta)), (theta, 0, pi))
plot_parametric((r*cos(theta), r*sin(theta)), (theta, 0, 64*pi))
```

```python
r1, r2 = symbols('r1 r2')
A = Integral(0.5*r2**2 - 0.5*r1**2, (theta,0,pi))
display(A)
r1 = 2 * (sin(theta)) **2
display ( Eq(S('r1'), r1, evaluate=False) )
r2 = 4 * (sin(theta)) **2
display ( Eq(S('r2'), r2, evaluate=False) )
A = Integral(0.5*r2**2 - 0.5*r1**2, (theta,0,pi))
display(A)
p1 = plot_parametric((r1*cos(theta), r1*sin(theta)), (theta, 0, pi), show = False)
p2 = plot_parametric((r2*cos(theta), r2*sin(theta)), (theta, 0, pi), show = False)
p1.extend(p2)
p1.show()
```

```python
r = symbols('r')
A = Integral(0.5 * r **2 ,(theta, 0, 2 * pi))
display(A)
r  = 1 + (1/2)*cos(6 * theta ) + (1/4)*sin(3 * theta)
display(r)
A = Integral(0.5 * r **2 ,(theta, 0, 2 * pi))
display(A)
display(A.doit())
display(A.doit().evalf())
plot_parametric((r*cos(theta), r*sin(theta)), (theta, 0, 2 * pi))
```
